```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
df_grouped = df.groupby(['Month', 'Day'])['Sales'].mean().unstack()
```

```python
fig = px.imshow(
    df_grouped,
    x=df_grouped.columns,y=df_grouped.index,
    title="Average Sales per Month and Day",
    labels={  # Optional: Customize axis labels
        "Month": "Month",
        "Day": "Day of Month",
        "color": "Average Sales"  # Colorbar label
    },color_continuous_scale='Viridis'
)
fig.show()
```

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
data = { 'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten(), }
```

```python
source = ColumnDataSource(data=data)
```

```python
p = figure()
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))
p.rect(x='x', y='y',width=1/10, height=1/10, source=source, fill_color=mapper, line_color=None)
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
